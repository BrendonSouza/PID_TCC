# Análise dos dados de velocidade dos motores

```python
import numpy as np
import matplotlib.pyplot as plt
import json

#le os dados dos arquivos de velocidades_direita.txt, atribui a uma variável as velocidades e a outra os tempos
with open('velocidades_direita.txt', 'r') as f:
    linesDireita = f.readlines()

velocidades_direita = []
tempos_direita = []

for line in linesDireita:
    data = json.loads(line)
    velocidades_direita.append(data['rpm'])
    tempos_direita.append(data['time'])

# converte todos os dados de velocidade para rpm
velocidades_direita = np.array(velocidades_direita)/6
velocidades_direita = np.array(velocidades_direita)/60

# plota o gráfico de velocidade em função do tempo

plt.plot(tempos_direita, velocidades_direita)
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (rps)')
plt.title('Velocidade da roda direita')

plt.show()

#corta os dados até o tempo de 1 segundo

```

```python

# Encontra o índice do tempo mais próximo de 1 segundo
indice_1_segundo = np.argmax(np.array(tempos_direita) >= 2.0)

# Corta os dados até o tempo de 1 segundo
tempos_direita_cortado = tempos_direita[:indice_1_segundo]
velocidades_direita_cortado = velocidades_direita[:indice_1_segundo]

# Plota o gráfico de velocidade em função do tempo
plt.plot(tempos_direita_cortado, velocidades_direita_cortado)
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (rps)')
plt.title('Velocidade da roda direita')

plt.show()
```

# Plota o gráfico da roda esquerda

```python


#le os dados dos arquivos de velocidades_direita.txt, atribui a uma variável as velocidades e a outra os tempos
with open('velocidades_esquerda.txt', 'r') as f:
    linesEsquerda = f.readlines()

velocidades_esquerda = []
tempos_esquerda = []

for line in linesEsquerda:
    data = json.loads(line)
    velocidades_esquerda.append(data['rpm'])
    tempos_esquerda.append(data['time'])

# converte todos os dados de velocidade para rpm
velocidades_esquerda = np.array(velocidades_esquerda)/6
velocidades_esquerda = np.array(velocidades_esquerda)/60

# calcula a media das velocidades
media_velocidades_esquerda = np.median(velocidades_esquerda)


# plota o gráfico de velocidade em função do tempo

plt.plot(tempos_esquerda, velocidades_esquerda)
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (rps)')
# traça a linha da meddiana
plt.axhline(y=media_velocidades_esquerda, color='r', linestyle='-', label='Mediana: '+str(media_velocidades_esquerda))
plt.title('Velocidade da roda esquerda')

# mostra a legenda, com o valor da mediana

plt.legend()


plt.show()



```

```python
# sobrep˜oe os gráficos de velocidade da roda direita e da roda esquerda

plt.plot(tempos_direita, velocidades_direita, label='Roda direita')
plt.plot(tempos_esquerda, velocidades_esquerda, label='Roda esquerda')
plt.xlabel('Tempo (s)')


# traça uma linha reta horizontal em 1.85 rps
plt.plot([0, 5], [1.88, 1.88], label='Velocidade de referência')
plt.ylabel('Velocidade (rps)')
plt.title('Velocidade das rodas')
plt.legend()

plt.show()
```

```python

# Encontra o índice do tempo mais próximo de 1 segundo
indice_1_segundo = np.argmax(np.array(tempos_direita) >= 2.0)

# Corta os dados até o tempo de 1 segundo
tempos_esquerda_cortado = tempos_esquerda[:indice_1_segundo]
velocidades_esquerda_cortado = velocidades_esquerda[:indice_1_segundo]

# Plota o gráfico de velocidade em função do tempo
plt.plot(tempos_esquerda_cortado, velocidades_esquerda_cortado)
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (rps)')
plt.title('Velocidade da roda esquerda')

plt.show()
```

```python
# sobrep˜oe os gráficos de velocidade da roda direita e da roda esquerda cortados

plt.plot(tempos_direita_cortado, velocidades_direita_cortado, label='Roda direita')
plt.plot(tempos_esquerda_cortado, velocidades_esquerda_cortado, label='Roda esquerda')
plt.xlabel('Tempo (s)')
# linha ideal, horizontal em 1.8 rps
plt.plot([0, 2], [1.85, 1.85], label='Ideal')
plt.ylabel('Velocidade (rps)')
plt.title('Velocidade das rodas')
plt.legend()

plt.show()
```

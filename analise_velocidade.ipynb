# An√°lise dos dados de velocidade dos motores

```python
import numpy as np
import matplotlib.pyplot as plt
import json

#le os dados dos arquivos de velocidades_direita.txt, atribui a uma vari√°vel as velocidades e a outra os tempos
with open('velocidades_direita.txt', 'r') as f:
    linesDireita = f.readlines()

velocidades_direita = []
tempos_direita = []

for line in linesDireita:
    data = json.loads(line)
    velocidades_direita.append(data['rpm'])
    tempos_direita.append(data['time'])


velocidade_direita_deg = np.array(velocidades_direita)
mediana_deg = np.median(velocidade_direita_deg)
# converte todos os dados de velocidade para rpm
velocidades_direita = np.array(velocidades_direita)/6
velocidades_direita = np.array(velocidades_direita)/60

mediana_rps = np.median(velocidades_direita)


# plota o gr√°fico de velocidade em fun√ß√£o do tempo

plt.plot(tempos_direita, velocidades_direita)
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (rps)')
plt.title('Velocidade da roda direita')

# plt mediana rps
plt.axhline(y=mediana_rps, color='r', linestyle='-', label='Mediana: '+str(mediana_rps))
plt.legend()
plt.show()



#Plota o grafico de velocidade_deg em fun√ß√£o do tempo

plt.plot(tempos_direita, velocidade_direita_deg)
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (deg/s)')
plt.title('Velocidade da roda direita')

plt.axhline(y=mediana_deg, color='r', linestyle='-', label='Mediana: '+str(mediana_deg))

plt.legend()

plt.show()
```

```python

# Encontra o √≠ndice do tempo mais pr√≥ximo de 1 segundo
indice_1_segundo = np.argmax(np.array(tempos_direita) >= 2.0)

# Corta os dados at√© o tempo de 1 segundo
tempos_direita_cortado = tempos_direita[:indice_1_segundo]
velocidades_direita_cortado = velocidades_direita[:indice_1_segundo]

# Plota o gr√°fico de velocidade em fun√ß√£o do tempo
plt.plot(tempos_direita_cortado, velocidades_direita_cortado)
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (rps)')
plt.title('Velocidade da roda direita')

plt.show()
```
